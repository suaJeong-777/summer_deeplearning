###파이썬의 클래스 
클래스 : 개발자가 직접 독자적으로 만드는 자료형으로 개발자가 클래스를 만들면 
그 클래스만의 전용 함수(메서드)와 속성을 정의할 수도 있다.

```python
#파이썬의 클래스
class 클래스 이름:
  def __init__(self, 인수, ....):  #생성자: 클래스를 초기화 하는 방법을 정의 
  ...
  def 메서드 이름1(self, 인수, ...): #메서드 1
  ...
  def 메서드 이름2(self, 인수, ...) #메서드 2
```

```python
#예제 1. man이라는 클래스를 만들어보자 
class Man:      #여기서 Man이라는 새로운 클래스를 정의함
  def __init__(self, name):   #name이라는 인수를 받고, 그 인수로 인스턴스 변수인 self.name을 초기화함. 이때 인스턴스 변수란 인스턴스 별로 저장되는 변수 
    self.name = name
    print("initialized!")
  
  def hello(self):
    print("Hello"+self.name+"!")

  def goodbye(self):
    print("Good-bye "+ self.name +"!")

m=Man("David")        #Man클래스에서 m이라는 인스턴스(객체)를 생성
m.hello()
m.goodbye()
```

###넘파이(numpy) :수치계산용 외부 라이브러리

```python
import numpy as np

x=np.array([1.0,2.0,3.0])
print(x)
print(type(x))

y=np.array([2.0,4.0,6.0])     
print(x+y)     #이때 배열 x와 y의 원소의 개수가 다르다면 오류가 발생한다. 
print(x-y)
print(x*y)
print(x/y)

#넘파이 배열은 원소별 계산 뿐 아니라 넘파이 배열과 수피 하나(스칼라값)의 조합으로 된 산술연산도 수행할 수 있다
#이 경우 스칼라값과의 계산이 넘파이 배열의 원소별로 한번씩 수행되고 이 기능을 브로드캐스트라고 함
x=np.array([1.0,2.0,3.0])
print(x/2.0)

#numpy의 n차원 배열
#예제1. numpy의 2차원 배열
A=np.array([[1,2],[3,4]])
print(A)
print(A.shape)
print(A.dtype)

#예제2. 행렬의 산술연산 : 형상이 같은 행렬끼리면 행렬의 산술연산도 대응하는 원소별로 계산됨
#1차원 배열은 벡터(vector), 2차원 배열은 행렬(matrix), 3차원 배열은 텐서(tensor) 라고 함
B=np.array([[3,0],[0,6]])
print(A+B)
print(A*B)

#브로드캐스트 : 형상이 다른 배열끼리의 연산에서 발생하는 기능
A=np.array([[1,2],[3,4]])
B=np.array([10,20])   #B의 배열이 연산 과정에서 자동으로 ([[10,20],[10,20]])의 형태로 변환되어 연산이 진행됨
print(A*B)

#원소 접근
X=np.array([[51,55],[14,19],[0,4]])
print(X)
print(X[0])  #0행
print(X[0][1]) #0행 1열
#또는 
for row in X: #for문으로도 접근할 수 있다
  print(row)

X=X.flatten()  #X를 1차원으로 변환(평탄화)
print(X)
print(X[X>15]) #X값이 15보다 큰 값들만 출력
print(X[np.array([0,2,4])]) #인텍스라 0,2,4인 원소 출력
```

###matplotlib : 그래프 그리기를 위한 외부 라이브러리 

```python
#그래프를 그리기 위해서는 matplotliv의 pyplot 모듈을 사용해야 함

##예제1.. sin함수 그려보기 
import numpy as np
import matplotlib.pyplot as plt
#데이터 준비 
x=np.arange(0,6,0.1)  #0부터 6까지 0.1 간격으로 생성
y=np.sin(x)
#그래프 그리기 
plt.plot(x,y)
plt.show()

##예제2. cos 함수 기능 추가
#데이터 준비 
x=np.arange(0,6,0.1)  #0부터 6까지 0.1 간격으로 생성
y1=np.sin(x)
y2=np.cos(x)
#그래프 그리기 
plt.plot(x,y1,label="sin")
plt.plot(x,y2,linestyle="--",label="cos") #cos함수는 점선으로 그리기
plt.xlabel="x" #x축의 이름
plt.ylabel="y" #y축의 이름
plt.title('sin&cos') #제목
plt.legend()  #범례 표시하기 (좌측하단의 작은 박스)
plt.show()
```
